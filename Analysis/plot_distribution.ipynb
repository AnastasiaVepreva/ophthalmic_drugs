```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv(r'C:\Users\Anastasia\Desktop\ocular drugs\Data\descriptors_120raws.csv')
```

```python
np.arange(0.5,7.5,0.5)
```

```python
def get_class(x):
    for val in np.arange(0.5,7.5,0.5):
        if (x < val) and (x > val - 0.5):
            return f'{val-0.5}-{val}'
```

```python
df.head()
```

```python
df['class'] = df['logPerm'].apply(get_class)
df
```

```python
pd.value_counts(df['class']).keys()
```

```python
val_counts = df['class'].value_counts()
classes = val_counts.keys()
numbers = val_counts
```

```python
plt.pie(numbers, labels = classes, autopct='%1.1f%%')
plt.show()
```
