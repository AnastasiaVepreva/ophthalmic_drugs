```python
df1 = pd.read_csv(r'C:\Users\Anastasia\Documents\GitHub\ophthalmic_drugs\Data\merged_reviews1.csv')
df1
```

```python
df2 = pd.read_csv(r'C:\Users\Anastasia\Documents\GitHub\ophthalmic_drugs\Data\Xuchun2002.csv')
df2['perm(cm/s)'] = df2['lopP'].apply(lambda s: s.replace ('?', '-'))
df2['perm(cm/s)'] = df2['perm(cm/s)'].astype(float)
df2['perm(cm/s)'] = df2['perm(cm/s)'].apply(lambda s: 10**s*10000000)
df2.insert(5, "loss", 0)
df2=df2[['SMILES', 'Name', 'perm(cm/s)', 'loss', 'DOI']]
df2
```

```python
df3 = pd.concat([df1, df2], ignore_index=True)
df3 = df3.drop_duplicates(subset='SMILES', keep='first')
df3.to_csv('merged_reviews2.csv', index=False)
```
