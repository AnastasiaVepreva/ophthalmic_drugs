```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostRegressor
```

```python
df=pd.read_csv(r'C:\Users\Anastasia\Desktop\ocular drugs\Data\descriptors_120raws.csv')
```

```python
df2 = df.loc[[16, 54, 87, 119, 15, 25, 30, 76, 88, 110, 0, 27, 72, 91, 73, 8, 23, 105, 92, 37, 51, 36, 13, 115, 104, 82, 5, 85, 66, 7]]
```

```python
df3 = df.drop (index=[16, 54, 87, 119, 15, 25, 30, 76, 88, 110, 0, 27, 72, 91, 73, 8, 23, 105, 92, 37, 51, 36, 13, 115, 104, 82, 5, 85, 66, 7])
```

```python
X_test = df2[df2.columns[:-1]]
y_test = df2[df2.columns[-1]]
X_train = df3[df3.columns[:-1]]
y_train = df3[df3.columns[-1]]
```

```python
regr = AdaBoostRegressor(random_state=0, n_estimators=10)
regr.fit(X_train, y_train)
regr.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':np.arange(5, 100, 5), 'loss':['linear', 'square', 'exponential'], 'learning_rate':np.logspace(-5, -1, num=10)}
adaboost = AdaBoostRegressor()
regr = GridSearchCV(adaboost, parameters)
regr.fit(X_train, y_train)
```
