```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Anastasia\Documents\GitHub\ophthalmic_drugs\Data\melanin\fingerprints_melanin.csv')
df
```

```python
def f(row):
    global val
    if row['Melanin binding Fraction unbound (%)'] < 1:
        val = 0
    else:
        val = 1
    return val


```

```python
df['Class'] = df.apply(f, axis=1)
```

```python
df
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
x = np.array(df.iloc[0:780, 1:168])
y = np.array(df[['Class']])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
reg = LazyClassifier(verbose=0, 
                    ignore_warnings=False, 
                    custom_metric=None)
models, predictions = reg.fit(xtrain, xtest, ytrain, ytest)
models
```

```python
from sklearn.ensemble import RandomForestClassifier	
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(xtrain, ytrain)
```

```python
clf.score(xtest, ytest)
```

```python
clf.score(xtrain, ytrain)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(random_state=50)
clf.fit(xtrain, ytrain)
```

```python
clf.score(xtest, ytest)
```

```python
clf.score(xtrain, ytrain)
```
