```python
import pandas as pd
import numpy as np

df1 = pd.read_csv(r'C:\Users\Anastasia\Documents\GitHub\ophthalmic_drugs\Data\Kidron2010.csv', usecols = ['SMILES', 'Name', 'perm(cm/s)', 'DOI'])
df1.head(5)
df1.insert(3, "loss", 0)
df1['perm(cm/s)'] = df1['perm(cm/s)'].apply(lambda s: s*10000000)

```

```python
df2 = pd.read_csv(r'C:\Users\Anastasia\Documents\GitHub\ophthalmic_drugs\Data\Ramsay2018.csv', sep=',', engine = 'python', encoding='unicode_escape', usecols = ['Smiles', 'Name', 'Papp (cm/s) * 10^-7', 'DOI'])
df2['perm(cm/s)'] = df2['Papp (cm/s) * 10^-7'].apply(lambda s: s.split('±')[0])
df2['loss'] = df2['Papp (cm/s) * 10^-7'].apply(lambda s: s.split('±')[1])
df2=df2[['Smiles', 'Name', 'perm(cm/s)', 'loss', 'DOI']]
df2 = df2.rename(columns={'Smiles': 'SMILES'})

```

```python
df3 = pd.read_csv(r'C:\Users\Anastasia\Documents\GitHub\ophthalmic_drugs\Data\Ronald1983.csv', sep=',', engine = 'python', encoding='unicode_escape', usecols = ['SMILES', 'Name', 'logPt, cm/c', 'DOI'])
df3.insert(3, "loss", 0)
df3['perm(cm/s)'] = df3['logPt, cm/c'].apply(lambda s: 10**s*10000000)
df3=df3[['SMILES', 'Name', 'perm(cm/s)', 'loss', 'DOI']]
df3.dropna ()

```

```python
df4 = pd.concat([df1, df2, df3], ignore_index=True)
df4 = df4.drop_duplicates(subset='SMILES', keep='first')
df4 = df4.dropna()
df4.to_csv('merged_reviews1.csv', index=False)
df4

```
